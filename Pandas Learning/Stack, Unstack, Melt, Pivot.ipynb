```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_gb = tips.groupby(["day", "sex", "smoker"]).agg({"size": "sum"})
tips_gb
```

# UNSTACK

```python
# Unstack moves row labels to column headers
tips_us = tips_gb.unstack()
tips_us
```

```python
tips_gb.unstack().unstack()
```

```python
tips_gb.unstack(0)
```

```python
tips_gb.unstack(1)
```

```python
tips_gb.unstack(2)
```

```python
tips_us.columns
```

```python
tips_us[[("size", "Yes")]]
```

```python
tips_us_copy = tips_us.copy()

tips_us_copy.columns = ["__".join(col).strip() for col in tips_us.columns.values]
```

```python
tips_us_copy
```

```python
tips_us_copy.size__No / tips_us_copy.size__Yes
```

```python
tips_us_copy_2 = tips_gb.unstack().unstack()

tips_us_copy_2.columns = ["__".join(col).strip() for col in tips_us_copy_2.columns.values]
```

```python
tips_us_copy_2
```

```python
tips_us_copy_2.size__No__Female / tips_us_copy_2.size__Yes__Female
```

# MELTING

```python
cheese = pd.DataFrame({"first" : ["John", "Mary"],
                      "last" : ["Doe", "Bo"],
                      "height":[5, 6],
                      "weight":[150, 200]})
cheese
```

```python
# melt does staking in one operation
cheese.melt(id_vars=["height", "weight"])
```

```python
cheese.melt(id_vars=["first", "last"])
```

```python
# It can be done with stacking in two steps
cheese_copy = cheese.copy()
cheese_copy.set_index(["first", "last"], inplace = True)
cheese_copy
```

```python
cheese_copy.stack().reset_index()
```
